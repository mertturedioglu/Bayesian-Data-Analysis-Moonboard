### Load Data and Standardize

```python
df = pd.read_csv("../dataset/benchmark_problems_featured.csv")
df.drop(["Unnamed: 0",
        "Grade_thermo",
        "Moves",
        "Holdsets",
         "holds",
        "lengths_of_moves",
        "Id",
        "numeric_grades",
        "Grade_ohe"],
       axis=1,
       inplace=True)
df.sort_values("Grade_numeric", inplace=True)
df.head(5)

# Standardization

df["fmd"] = (df["first_move_diff"] - df["first_move_diff"].mean()) / df["first_move_diff"].std()
df["lmd"] = (df["last_move_diff"] - df["last_move_diff"].mean()) / df["last_move_diff"].std()
df["n"] = (df["number_of_holds"] - df["number_of_holds"].mean()) / df["number_of_holds"].std()
df["mdh"] = (df["mean_distance_of_holds"] - df["mean_distance_of_holds"].mean()) / df["mean_distance_of_holds"].std()
df["yp"] = (df["yellow_percent"] - df["yellow_percent"].mean()) / df["yellow_percent"].std()
```

# DAG

```python
pgm = daft.PGM(line_width=1.3,)

# Grade
pgm.add_node("G","G",3,0, scale=2)
# First Move Difficulty
pgm.add_node("FMD","FMD",2,-1)
# Number of Holds
pgm.add_node("N","N",3,3)
# Mean Distance Between Holds
pgm.add_node("MDH","MDH",6,0)
# Yellow Percent
pgm.add_node("YP","YP",0,0)
# Last Move Difficulty
pgm.add_node("LMD","LMD",2,1)

pgm.add_edge("YP", "G")
pgm.add_edge("N", "YP")
pgm.add_edge("N", "G")
pgm.add_edge("N", "MDH")
pgm.add_edge("YP", "LMD")
pgm.add_edge("YP", "FMD")
pgm.add_edge("LMD", "G")
pgm.add_edge("MDH", "G")
pgm.add_edge("FMD", "G")
pgm.render()
```

### Move Difficulty Model:


$$
\begin{align*}
{G}_i \sim \text{OrderedLogit}(\phi_i, \alpha)
\\
\phi_i = \beta_{fmd}{FMD}_i + \beta_{lmd}{LMD}_i + \beta_{YP}{YP}_i + \beta_{N}{N}_i
\\
\beta_{fmd}, \beta_{lmd}, \beta_{YP}, \beta_{N} \sim Normal(0, 1)
\\
\alpha_j \sim \text{Normal}(0, 1)
\end{align*}
$$

```python
FMD = df.fmd.values
LMD = df.lmd.values
YP = df.yp.values
N = df.n.values

G = df.Grade_numeric.values
```

```python
with pm.Model() as mMD:
    bFMD = pm.Normal('bFMD', 0, 1)
    bLMD = pm.Normal('bLMD', 0, 1)
    bYP = pm.Normal('bYP', 0, 1)
    bN = pm.Normal('bN', 0, 1)
    
    # define data with pm.Data to be able to modify it later
    FMD = pm.Data("FMD", FMD)
    LMD = pm.Data("LMD", LMD)
    YP = pm.Data("YP", YP)
    N = pm.Data("N", N)

    alpha = pm.Normal("alpha", 0.0, 1, transform = pm.distributions.transforms.ordered, shape = 11, initval=np.arange(11) - 5.5,)
    phi = pm.Deterministic("phi", bFMD * FMD + bLMD * LMD + bYP * YP + bN * N)
    grade = pm.OrderedLogistic('grade', eta = phi, cutpoints = alpha, observed = G)
    trace_mMD = pm.sample(2000, return_inferencedata=True, tune=2000, chains=4, target_accept=0.95)
```

```python
az.summary(trace_mMD, var_names=["alpha", "bFMD", "bLMD", "bYP", "bN"])
```

```python
az.plot_trace(trace_mMD, compact=False, var_names=["alpha", "bFMD", "bLMD", "bYP", "bN"])
```

```python
az.plot_posterior(trace_mMD ,hdi_prob=0.92, var_names=["alpha","bFMD", "bLMD", "bYP", "bN"])
```

### Cases

```python
# fmd = first move difficulty
# lmd = last move difficulty

number_of_cases= 20

fmd_linspaced = np.linspace(-0.1,1, num=number_of_cases)
lmd_linspaced = np.linspace(-0.5, 10, num=number_of_cases)
n_linspaced = np.linspace(3.5, -2.5, num=number_of_cases)
yp_linspaced = np.linspace(-0.5, 3, num =number_of_cases)

cases = []
for i, j, k, l in zip(fmd_linspaced, lmd_linspaced, n_linspaced, yp_linspaced):
    cases.append([i, j, k, l])

cases_df = pd.DataFrame(np.array(cases), columns=["FMD","LMD", "N", "YP"])
display(cases_df)
```

```python
# Posterior predictives

with mMD:
    pm.set_data({
        "FMD": cases_df.FMD.values,
        "LMD": cases_df.LMD.values,
        "N": cases_df.N.values,
        "YP": cases_df.YP.values,
    })
    pp_mMD = pm.sample_posterior_predictive(trace_mMD, var_names = ["grade"])
```

```python
def drawHistforPost(prediction, cases, index):
    state = cases.loc[index]
    postdata = prediction.posterior_predictive.sel(grade_dim_0 = index)
    postdatas = postdata.stack(sample=("chain", "draw"))

    # Histogram settings
    bins, xticks, xlabels = (
    np.arange(13) - 0.5,
    np.arange(12),
    list(df["Grade"].unique())
    )

    # Histogram
    _, axes = plt.subplots()
    axes.hist(postdatas.grade, bins=bins, rwidth = 0.5, color = "darkred", alpha = 0.7)
    axes.set_ylabel("frequency")
    axes.set_xlabel("Grade")
    axes.set_xticks(xticks)
    axes.set_xticklabels(xlabels)
    axes.set_title(f"fmd = {state.FMD}, lmd = {state.LMD}, n = {state.N}, yp = {state.YP}")
    return axes
```

```python
for i in range(len(cases_df)):
    drawHistforPost(pp_mMD, cases_df, i)
    plt.show()
```

### Mean Distance - Number of holds - Yellow Percent Model

$$
\begin{align*}
{G}_i \sim \text{OrderedLogit}(\phi_i, \alpha)
\\
\phi_i = \beta_{N}{N}_i + \beta_{MDH}{MDH}_i + \beta_{YP}{YP}_i
\\
\beta_{N}, \beta_{MDH}, \beta_{YP} \sim Normal(0, 1)
\\
\alpha_j \sim \text{Normal}(0, 1)
\end{align*}
$$

```python
N = df.n.values
MDH = df.mdh.values
YP = df.yp.values

G = df.Grade_numeric.values
```

```python
with pm.Model() as mMDHNYP:
    bN = pm.Normal('bN', 0, 1)
    bMDH = pm.Normal('bMDH', 0, 1)
    bYP = pm.Normal('bYP', 0, 1)

    # define data with pm.Data to be able to modify it later
    N = pm.Data("N", N)
    MDH = pm.Data("MDH", MDH)
    YP = pm.Data("YP", YP)

    alpha = pm.Normal("alpha", 0.0, 1, transform = pm.distributions.transforms.ordered, shape = 11, initval=np.arange(11) - 5.5,)
    phi = pm.Deterministic("phi", bN * N + bMDH * MDH + bYP * YP)
    grade = pm.OrderedLogistic('grade', eta = phi, cutpoints = alpha, observed = G)
    trace_mMDHNYP = pm.sample(2000, return_inferencedata=True, tune=2000, chains=4, target_accept=0.95)
```

```python
az.summary(trace_mMDHNYP, var_names=["alpha","bMDH", "bYP", "bN"])
```

```python
az.plot_trace(trace_mMDHNYP, compact=False, var_names=["alpha","bMDH", "bYP", "bN"])
```

```python
az.plot_posterior(trace_mMDHNYP ,hdi_prob=0.92, var_names=["alpha","bN","bMDH", "bYP"])
```

```python
def drawHistforPost(prediction, cases, index):
    state = cases.loc[index]
    postdata = prediction.posterior_predictive.sel(grade_dim_0 = index)
    postdatas = postdata.stack(sample=("chain", "draw"))

    # Histogram settings
    bins, xticks, xlabels = (
    np.arange(13) - 0.5,
    np.arange(12),
    list(df["Grade"].unique())
    )

    # Histogram
    _, axes = plt.subplots()
    axes.hist(postdatas.grade, bins=bins, rwidth = 0.5, color = "darkred", alpha = 0.7)
    axes.set_ylabel("frequency")
    axes.set_xlabel("Grade")
    axes.set_xticks(xticks)
    axes.set_xticklabels(xlabels)
    axes.set_title(f"n = {state.N}, mdh = {state.MDH}, yp={state.YP}")
    return axes
```

### Load Data and Standardize

```python
df = pd.read_csv("../dataset/benchmark_problems_featured.csv")
df.drop(["Unnamed: 0",
        "Grade_thermo",
        "Moves",
        "Holdsets",
         "holds",
        "lengths_of_moves",
        "Id",
        "numeric_grades",
        "Grade_ohe"],
       axis=1,
       inplace=True)
df.sort_values("Grade_numeric", inplace=True)
df.head(5)

# Standardization

df["fmd"] = (df["first_move_diff"] - df["first_move_diff"].mean()) / df["first_move_diff"].std()
df["lmd"] = (df["last_move_diff"] - df["last_move_diff"].mean()) / df["last_move_diff"].std()
df["n"] = (df["number_of_holds"] - df["number_of_holds"].mean()) / df["number_of_holds"].std()
df["mdh"] = (df["mean_distance_of_holds"] - df["mean_distance_of_holds"].mean()) / df["mean_distance_of_holds"].std()
```

# DAG

```python
pgm = daft.PGM()
# First Move Difficulty
pgm.add_node("FMD","FMD",0,2)
# Last Move Difficulty
pgm.add_node("LMD","LMD",2,2)
# Grade
pgm.add_node("G","G",1,1)

pgm.add_edge("LMD", "G")
pgm.add_edge("FMD", "G")
pgm.render()
```

### Move Difficulty Model:


$$
\begin{align*}
{G}_i \sim \text{OrderedLogit}(\phi_i, \alpha)
\\
\phi_i = \beta_{fmd}{FMD}_i + \beta_{lmd}{LMD}_i
\\
\beta_{fmd}, \beta_{lmd} \sim Normal(0, 0.5)
\\
\alpha_j \sim \text{Normal}(0, 1)
\end{align*}
$$

```python
FMD = df.fmd.values
LMD = df.lmd.values
G = df.Grade_numeric.values
```

```python
with pm.Model() as mMD:
    bFMD = pm.Normal('bFMD', 0, 0.5)
    bLMD = pm.Normal('bLMD', 0, 0.5)

    # define data with pm.Data to be able to modify it later
    FMD = pm.Data("FMD", FMD)
    LMD = pm.Data("LMD", LMD)

    alpha = pm.Normal("alpha", 0.0, 1, transform = pm.distributions.transforms.ordered, shape = 11, initval=np.arange(11) - 5.5,)
    phi = pm.Deterministic("phi", bFMD * FMD + bLMD * LMD)
    grade = pm.OrderedLogistic('grade', eta = phi, cutpoints = alpha, observed = G)
    trace_mMD = pm.sample(return_inferencedata=True, tune=2000, chains=4, target_accept=0.95)
```

```python
az.summary(trace_mMD, var_names=["alpha"])
```

```python
az.plot_trace(trace_mMD, compact=False)
```

```python
az.plot_posterior(trace_mMD ,hdi_prob=0.92, var_names=["alpha","bFMD","bLMD"])
```

### Posterior predictive simulations - move difficulty model

We generate posterior predictions for the cases below:

#### Cases

```python
# fmd = first move difficulty
# lmd = last move difficulty

n= 20

fmd_linspaced = np.linspace(-0.1,1, num=n)
lmd_linspaced = np.linspace(-0.5, 10, num=n)

cases = []
for i, j in zip(fmd_linspaced, lmd_linspaced):
    cases.append([i,j])

cases_df = pd.DataFrame(np.array(cases), columns=["FMD","LMD"])
display(cases_df)
```

```python
# Posterior predictives

with mMD:
    pm.set_data({
        "FMD": cases_df.FMD.values,
        "LMD": cases_df.LMD.values
    })
    pp_mMD = pm.sample_posterior_predictive(trace_mMD, var_names = ["grade"])
```

```python
def drawHistforPost(prediction, cases, index):
    state = cases.loc[index]
    postdata = prediction.posterior_predictive.sel(grade_dim_0 = index)
    postdatas = postdata.stack(sample=("chain", "draw"))

    # Histogram settings
    bins, xticks, xlabels = (
    np.arange(13) - 0.5,
    np.arange(12),
    list(df["Grade"].unique())
    )

    # Histogram
    _, axes = plt.subplots()
    axes.hist(postdatas.grade, bins=bins, rwidth = 0.5, color = "darkred", alpha = 0.7)
    axes.set_ylabel("frequency")
    axes.set_xlabel("Grade")
    axes.set_xticks(xticks)
    axes.set_xticklabels(xlabels)
    axes.set_title(f"fmd = {state.FMD}, lmd = {state.LMD}")
    return axes
```

```python
for i in range(len(cases_df)):
    drawHistforPost(pp_mMD, cases_df, i)
    plt.show()
```

### Mean Distance - Number of holds Model

```python
pgm = daft.PGM()
# First Move Difficulty
pgm.add_node("N","N",0,2)
# Last Move Difficulty
pgm.add_node("MDH","MDH",2,2)
# Grade
pgm.add_node("G","G",1,1)

pgm.add_edge("N", "G")
pgm.add_edge("MDH", "G")
pgm.add_edge("N", "MDH")
pgm.render() 
```

$$
\begin{align*}
{G}_i \sim \text{OrderedLogit}(\phi_i, \alpha)
\\
\phi_i = \beta_{N}{N}_i + \beta_{MDH}{MDH}_i
\\
\beta_{N}, \beta_{MDH} \sim Normal(0, 0.5)
\\
\alpha_j \sim \text{Normal}(0, 1)
\end{align*}
$$

```python
N = df.n.values
MDH = df.mdh.values
G = df.Grade_numeric.values
```

```python
with pm.Model() as mMDHN:
    bN = pm.Normal('bN', 0, 0.5)
    bMDH = pm.Normal('bMDH', 0, 0.5)

    # define data with pm.Data to be able to modify it later
    N = pm.Data("N", N)
    MDH = pm.Data("MDH", MDH)

    alpha = pm.Normal("alpha", 0.0, 1, transform = pm.distributions.transforms.ordered, shape = 11, initval=np.arange(11) - 5.5,)
    phi = pm.Deterministic("phi", bN * N + bMDH * MDH)
    grade = pm.OrderedLogistic('grade', eta = phi, cutpoints = alpha, observed = G)
    trace_mMDHN = pm.sample(return_inferencedata=True, tune=2000, chains=4, target_accept=0.95)
```

```python
az.summary(trace_mMDHN)
```

```python
az.plot_trace(trace_mMDHN, compact=False)
```

```python
az.plot_posterior(trace_mMDHN ,hdi_prob=0.92, var_names=["alpha","bN","bMDH"])
```
